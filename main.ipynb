```python
# Precisamos apenas de títulos com a região 'US'

selecao = df['region'] == 'US'
```

```python
# Criando um novo dataframe com a região 'US'

df_us = df[selecao]
```

```python
# Existem títulos duplicados, mas o campo 'ordering' é importante. Vamos ordenar esta coluna e depois manter 
# apenas a primeira seleção dos duplicados
df_us = df_us.sort_values(['titleId', 'ordering'], ascending=[True, True])
df_us.head(50)
```

```python
# Mantendo os primeiros itens duplicados e removendo os outros.
df_us = df_us.drop_duplicates(subset='titleId', keep='first')
df_us.head(50)
```

```python
#Realizei um join dos data.frames que constavam ano e titulo do filme(US)

df_us_year = pd.merge(df_year, df_us, left_on='tconst', right_on='titleId', how='outer')
```

```python
df_us_year.head()
```

```python
#retirando os titulos nulos, possiveis filmes estrangeiros nao regionalizados para US
df_us_year = df_us_year.dropna(subset=['title'])
```

```python
df_us_year.head(50)
```

```python
df_us_year.info()
```

```python
#Juntando o ultimo dataset com o campo nota/rating
df_imdb = pd.merge(df_us_year, df_imdb_rat, on='tconst', how='outer')
```

```python
df_imdb.head(15)
```

```python
df_imdb.info()
```

```python
0 == 0
```

```python
#precisei fazer essa alteração, pois havia valores nulos que não em formato str. pra que essa função funcione, não pode haver palavras nas celulas.
df_imdb['endYear'] = df_imdb['endYear'].replace([r'\N'], '0')
df_imdb['startYear'] = df_imdb['startYear'].replace([r'\N'], '0')
```

```python
# tranformando o campo em numérico
df_imdb['endYear'] = pd.to_numeric(df_imdb['endYear'])
df_imdb['startYear'] = pd.to_numeric(df_imdb['startYear'])
```

```python
df_imdb.info()
```
