```python
#DATASET IMDB - RATING
df_imdb_rat = pd.read_csv('./imdb_ratings.tsv', sep='\t', encoding='UTF-8')
df_imdb_rat.head()
```

```python
#DATASET IMDB COM OS TÍTULOS DOS FILMES

df_imdb_titles = pd.read_csv('./imdb_titles.tsv', sep='\t', encoding='UTF-8')
df_imdb_titles.head()
```

```python
df = df_imdb_titles.copy()
```

```python
df.shape[0]
```

```python
# Precisamos apenas de títulos com a região 'US'

selecao = df['region'] == 'US'
```

```python
#reduzimos de 22 milhoes de filmes para 1M
df[selecao].shape[0]
```

```python
# Criando um novo dataframe com a região 'US'

df_us = df[selecao]
```

```python
df_us.head(50)
```

```python
# Existem títulos duplicados, mas o campo 'ordering' é importante. Vamos ordenar esta coluna e depois manter 
# apenas a primeira seleção dos duplicados
df_us = df_us.sort_values(['titleId', 'ordering'], ascending=[True, True])
df_us.head(50)
```

```python
# Mantendo os primeiros itens duplicados e removendo os outros.
df_us = df_us.drop_duplicates(subset='titleId', keep='first')
df_us.head(50)
```

```python
# Verificando a situação da existência de valores únicos no 'titleId'
df_us['titleId'].is_unique
```

```python
df_us['title'].is_unique
```

```python
#df_us.tail() - pegando as ultimas linhas
#df_us.sample(30)- pega um exemplo aleatorio de linhas
```

```python
#verificando os titulos duplicados, nao sao o mesmo filme, devem ser refilmagens
df_us[df_us.duplicated(subset='title')].head(50)
```

```python
# Recortar as colunas necessarias
df_us = df_us[['titleId', 'title']]
df_us.head()
```

```python
# Titulos para lower case 
df_us['title'] = df_us.title.str.lower()
df_us.sample(50)
```

```python

df_us.info()
```

```python
# Adicionando outra coluna com o ano de produção;
# Novo dataframe do IMDB com o ano

df_2 = pd.read_csv('./imdb2.tsv', sep='\t', encoding='UTF-8')
df_2.head()
```

```python
# Um novo DF com apenas o Identificador 'tconst' e o ano

df_year = df_2[['tconst', 'startYear']].copy()
```

```python
df_year.info()
```

```python
df_year.head()
```

```python
df_year.info()
```

```python
#modificando o campo nulo para '0' 
#df_year['endYear'] = df_year['endYear'].replace([r'\N'], '0')
df_year['startYear'] = df_year['startYear'].replace([r'\N'], '0')
# tranformando o campo em numérico
#df_year['endYear'] = pd.to_numeric(df_year['endYear'])
df_year['startYear'] = pd.to_numeric(df_year['startYear'])
```

```python
df_year.info()
```

```python
#Recortando filmes a partir de 1942
selecao = (df_year.startYear >= 1942) & (df_year.startYear < 2020)
df_year = df_year[selecao]
df_year.head()
```
